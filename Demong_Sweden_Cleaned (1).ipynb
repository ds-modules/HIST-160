```python
import pandas as pd
import matplotlib.pyplot as plt

Sweden = pd.read_csv('/Users/anushamohan/Desktop/DemogTrans_Sweden_1749to2015_be0101tab9utveng.csv')
Sweden.columns = Sweden.iloc[0]
Sweden = Sweden.reindex(Sweden.index.drop(0))
Sweden = Sweden.dropna(how='all', axis=1)
Sweden = Sweden.reindex(Sweden.index.drop(1))
Sweden = Sweden.head(266)
Sweden
```

```python
x_values_df = Sweden[Sweden[' Year'].notnull()]
x_values = x_values_df[' Year']
y_values_df = Sweden[Sweden['Live Births'].notnull()]
y_values = y_values_df['Live Births']
y_values_2_df = Sweden[Sweden['Deaths'].notnull()]
y_values_2 = y_values_2_df['Deaths']
plt.plot(x_values, y_values)
plt.plot(x_values, y_values_2)
plt.show()
```
